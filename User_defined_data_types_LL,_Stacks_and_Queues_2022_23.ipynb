## Linked List
Linked lists contains nodes. Each node contains data part and address part. Nodes data is saved in data part and address part consists of address of next node. If there is only one address part in the node then it is **Single Linked List (SLL)**.  Address part of I node consists of address of II node. Address part of II node consists of address of III node so on so forth. We can only move forward in SLL. We should not lose the address of first node. 

**Double Linked List (DLL)** consists of data part and two address parts, one address part holding the address of previous node and another address part holding the address of next node. As a node consists of the address of previous node as well as next node we can move both forward and backward. We shoudl not lose the address of first node.

Some of the Standard operations are:
1.   insert at the beginning of the list
2.   insert at the end of the list
3. delete the first node of the list
4. delete the last node of the list
5. print the list
6. reverse the list
7. Finding an element in a list
8. Print the head node or first node of the list
9. print the last node or end node of the list
10. Sort the list
11. insert element at nth index
12. insert element after or before an element
13. delete an element
14. delete element at nth node etc

## Stack
Working principle of stack is **Last in First Out**. It has one end open. Both insertions and deletions are done from the same end

Standard operations are:

1.   push(ele) - to insert an element into stack
2.   pop() - to remove element from stack
3. isEmpty() - returns True if the stack is empty and False if stack is not empty
4. tos() - prints element pointed to by **top of the stack** or the element that is inserted at the last

## Queue
Working principle of Queue is **First in First Out**.It has two ends open a. rear and b. front

Insertions are done at rear end and Deletions are done at front end

Standard operations are:

1.   Enqueue(ele) - to insert an element into Queue
2.   Dequeue() - to remove element from Queue
3. isEmpty() - returns True if the queue is empty and False if queue is not empty
4. rear() - prints element pointed to by rear in the queue or the last element inserted in the queue
5. front() - prints element pointed to by front in the queue or the element that is inserted first in Queue
6. printQueue() - prints the elements of the Queue

```python
#Queue  implementation
Q = []     #Empty list to implement Queue
def enqueue():
  ele=int(input("Enter element to insert in Queue"))
  Q.append(ele)

def dequeue():
  if isEmpty():
    print("Deletion not possible Q underflow")
  else:
    ele=Q.pop(0)
    print("Oldest element inserted is removed from Queue", ele)

def isEmpty():
  if len(Q)==0:
    return True
  else:
    return False

def front():
  if len(Q)==0:
    print("Queue underflow")
  else:
    print("Element pointed to by front or the oldest element in the Queue is ", Q[0])

def rear():
  if len(Q)==0:
    print("Queue underflow")
  else:  
    print("Element pointed to by rear or the last element inserted is ", Q[len(Q)-1])

def printQueue():
  if len(Q)==0:
    print("Queue underflow")
  else:
    print("Queue contents are ....")
    print(Q)

while True:
  print("1. Enqueue()\n2. Dequeue()\n3.isEmpty()\n4.front()\n5.rear\n6.print Queue\n7.Exit")
  ch=int(input("Enter your choice:  "))
  if ch==1:
    enqueue()
  elif ch==2:
    dequeue()
  elif ch==3:
    res=isEmtpy()
    if res:
      print("Queue is empty")
    else:
      print("Queue is not empty")
  elif ch==4:
    front()
  elif ch==5:
    rear()
  elif ch==6:
    printQueue()
  else:
    break
```
