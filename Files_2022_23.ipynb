A file is simply a named locaiton on the secondary storage to store related information. Depending on the way a file is opened for processing, a file can be classified as a text file or binary file

#Text File
Data is stored using human readable characters
Each line of data sends with a new line character
There is a special character called end-of-file (EOF) marker at the end of the file

#Binary File
Data is stored in the same format as it is stored in memory
There are no lines or newline character
There is an endo-of-file marker at the end of the file, which is usually implementation dependent

Text file Vs binary file - file system.jpeg

Note: Text files are easy to handle but may not be efficient in terms of storage space used. Binary files are complex but are more efficient in terms of storage space used. In some cases binary file is the only choice such as images

**Steps in processing a file**


*   Opening a file - open the file in an appropriate more
*   Reading from or writing onto a file - once file is opened successfully, data can be read from a file or written to a file in a variety of ways
*   Closing the file - Once processing is done, each file must be closed

**File object attributes**
Once a file is successfully opened, a *file object* is returned.
File object can be used to access the file 

*   closed - returns true if the file is closed and false otherwise
*   mode - returns file opening mode with which file is opened
*   name - returns name of the file

```python
#Program that opens a file and print its attributes
fo = open("test.txt", 'w')
print("File name is ", fo.name)
print("File closed", fo.closed)
print("File is opened in ", fo.mode, " mode")
```

**Possible trouble with opening a file**

* Trying to open a file for reading that does not exist   
* Trying to open a file for writing, nbut sufficient space is not available on the disk
*   Trying to open a file that has been corrupted





**Closing a file:**

When a program has finished with reading/writing of a file, it must be closed.

This task is carried out by using the close() function

Closing a file explicitly has two efects. 

*   Any data remaining in the buffer is writen to the file. It may be noted that the buffer used is invisibel to the programmer
*   Frees the communication areas used by the particular file so taht they are available for other files. These areas include the file object and the buffer *itself*



**The need for buffer:**
A buffer is a temporary storage area in main memory.It is necessary even if it is invisible
*   Scenario#1: suppose our program accesses the disk to read/write one character at a time. The disk system takes some tome to position the head on the correct track and then further waits for the correct secgtor of the track to come around. For every read/write operation, the motor that rotates the disk actually has to start from astand still. If you type several characters repeatedly and each one requires a separate disk access some of the characters would probably be lost. So this approach will be inefficient. To solve this problem buffer is used
*   Scenario#2: When a program write data to a file, it is actually stored in a buffer rather than being written immediately to the disk.  When the buffer is full, its contents are written to the disk all at once, and subsequent writes continue onto the buffer. Similarly when a program reads data from a file, a block of data, equal to the size of the buffer except the last block, is read from the disk file and is placed in the buffer. The input functions subsequently read data from a buffer. When a program attempts to read a data item that is not in buffer,  it forces to read another block of data containing the requested data item

```python
#program to demonstrate the usage of tell() and seek()
fo=open("data.txt"."wt")
fo.write("Hello world, how are you? hope you are practicing Python and you enjoyed thoroughly")
fo.close()

#open file in input mode
fo=open("data.txt","r")
print("Name of the file: ", fo.name)
s1=fo.read(10)
print("First 10 bytes of the file are",s1)
pos=fo.tell()
print("Current position",pos)

s2=fo.read(10)
print("Next 10 bytes of the file are",s2)
pos=fo.tell()
print("Current position",pos)

print("Repositioning file pointer to the beginning of file")
pos=fo.seek(0,0)
print("Current position", pos)
s3=fo.read(25)
print("First 25 bytes of the file are",s3)
pos=fo.tell()
print("Current position",pos)

fo.close()
```

```python
#Program to find the size of a given file. Name of the file is given by user
fn=input("Enter name of the file")
fo=open(fn,"r")
fo.seek(-0,2)    #Position the file pointer at the end of file
pos=fo.tell()
print("File size is",pos," bytes")
fo.close()
```

```python
#Program to read text from keyboard and write into a file
fo=open("/content/test3.txt","w")

while True:
  print("Type a line of text")
  line=input()
  fo.write(line+"\n")
  print("More lines Y/N",end='')
  yn=input()
  if yn not in ['y','Y'] :
    break
fo.close()

fo=open("/content/test3.txt","r")
print("Contents of file are")
txt = fo.readlines()

for line in txt:
  print(line, end='')
fo.close()
```

```python
#Program to copy the contents of one text file to another file libe-by-line
fname1=input("Enter source file name")
fname2=input("Enter destination file name")

fo1=open(fname1,"r")
fo2=open(fname2,"w")

for line in fo1:
  fo2.write(line)

fo1.close()
fo2.close()
```
