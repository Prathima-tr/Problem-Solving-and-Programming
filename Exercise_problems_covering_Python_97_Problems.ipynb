```python
# Q 3. Illustrate Arithmetic Operators
# The arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication and division.
# The following lists all the arithmetic operators supported by Python.

# i. Addition Operator (+)
# The addition operator adds two operands and the result is assigned to the left operand (left-side of = sign).
# Syntax: a + b
# Example:
a = 10
b = 20.2
c = a + b
# Output: Addition of a and b is 30.2
print(f"Addition of {a = } and {b = } is", c)
# The result is a floating point number because one of the operands is a floating point number.

# ii. Subtraction Operator (-)
# The subtraction operator subtracts the right operand from the left operand and the result is assigned to the left operand (left-side of = sign).
# Syntax: a - b
# Example:
a = 10
b = 20.2
c = a - b
# Output: Subtraction of a and b is -10.2
print(f"Subtraction of {a = } and {b = } is", c)
# The result is a floating point number because one of the operands is a floating point number.

# iii. Multiplication Operator (*)
# The multiplication operator multiplies two operands and the result is assigned to the left operand (left-side of = sign).
# Syntax: a * b
# Example:
a = 10
b = 20 + 3j
c = a * b
# Output: Multiplication of a and b is (200+30j)
print(f"Multiplication of {a = } and {b = } is", c)
# The result is a complex number because one of the operands is a complex number.

# iv. Division Operator (/)
# The division operator divides the left operand by the right operand. The result is assigned to the left operand (left-side of = sign).
# Syntax: a / b
# Example:
a = 10
b = 5
c = a / b
# Output: Division of a and b is 2.0
print(f"Division of {a = } and {b = } is", c)
# The result of division is always a floating point number.

# v. Floor Division Operator (//)
# The floor division operator divides the left operand by the right operand and returns the floor value of the result. The result is assigned to the left operand (left-side of = sign).
# Syntax: a // b
# Example:
a = 12
b = 5
c = a // b
# Output: Floor Division of a and b is 2
print(f"Floor Division of {a = } and {b = } is", c)
# The result of floor division is always an integer.

# vi. Modulus Operator (%)
# The modulus operator returns the remainder when the left operand is divided by the right operand. The result is assigned to the left operand (left-side of = sign).
# Syntax: a % b
# Example:
a = 12
b = 5
c = a % b
print(f"Modulus of {a = } and {b = } is", c)  # Output: Modulus of a and b is 2
# The result of modulus is always an integer.

# vii. Exponent Operator (**)
# The exponent operator raises the left operand to the power of the right operand. The result is assigned to the left operand (left-side of = sign).
# Syntax: a ** b
# Example:
a = 2
b = 3
c = a ** b
# Output: Exponent of a and b is 8
print(f"Exponent of {a = } and {b = } is", c)
# Output: Exponent of a and b is 8
print(f"Exponent of a = 4 and b = -1 is", 4 ** -1)
```

```python
# Q 4. Illustrate Logical Operators
# The logical operators are used to perform logical operations on boolean values. 
# The following lists all the logical operators supported by Python.

# i. AND Operator (and)
# The and operator returns True if both the operands are True. Otherwise, it returns False.
# Syntax: a and b
# Example:
a = True
b = False
c = a and b
print(f"AND of {a = }, {b = } is", c) # Output: AND of a, b is False
# For multiple operands, the and operator returns True if all the operands are True. Otherwise, it returns False.
# Example:
a = True
b = True
c = True
d = a and b and c
print(f"AND of {a = }, {b = }, {c = } is", d) # Output: AND of a, b, c is True

# ii. OR Operator (or)
# The or operator returns True if either of the operands is True. Otherwise, it returns False.
# Syntax: a or b
# Example:
a = True
b = False
c = a or b
print(f"OR of {a = }, {b = } is", c) # Output: OR of a, b is True
# For multiple operands, the or operator returns True if any of the operands is True. Otherwise, it returns False.
# Example:
a = False
b = False
c = False
d = a or b or c
print(f"OR of {a = }, {b = }, {c = } is", d) # Output: OR of a, b, c is False

# iii. NOT Operator (not)
# The not operator returns the opposite of the boolean value of the operand.
# Syntax: not a
# Example:
a = True
b = not a
print(f"NOT of {a = } is", b) # Output: NOT of a is False

# NOTE: The not operator is a unary operator. It takes only one operand. It cannot be used with multiple operands.

# NOTE: Short Circuit Evaluation
# The and and or operators are also known as short-circuit operators.
# They do not evaluate the second operand if the result can be determined by the first operand. 
# For example, if the first operand of the and operator is False, the second operand is not evaluated. 
# Similarly, if the first operand of the or operator is True, the second operand is not evaluated.
# Example:
a = 0 and 10 / 0
print(f"Short Circuit Evaluation of 0 and 10 / 0 is", a) # Output: Short Circuit Evaluation of 0 and 10 / 0 is 0
# In the above example, the second operand is not evaluated because the first operand is False.
# Example:
a = 10 or 10 / 0
print(f"Short Circuit Evaluation of 10 or 10 / 0 is", a) # Output: Short Circuit Evaluation of 10 or 10 / 0 is 10
# In the above example, the second operand is not evaluated because the first operand is True.

# These short-circuit operators are very useful in Python programming. They are used to avoid a few exceptions (like ZeroDivisionError in above case) in Python.
```

```python

# Q 5. Illustrate Relational Operators
# The relational operators are used to compare two operands.
# The following lists all the relational operators supported by Python.

# i. Equal To Operator (==)
# The equal to operator returns True if the two operands are equal. Otherwise, it returns False.
# Syntax: a == b
# Example:
a = 10
b = 10
c = a == b
# Output: Equal To of a and b is True
print(f"Equal To of {a = } and {b = } is", c)

# ii. Not Equal To Operator (!=)
# The not equal to operator returns True if the two operands are not equal. Otherwise, it returns False.
# Syntax: a != b
# Example:
a = 10
b = 10
c = a != b
# Output: Not Equal To of a and b is False
print(f"Not Equal To of {a = } and {b = } is", c)

# iii. Greater Than Operator (>)
# The greater than operator returns True if the left operand is greater than the right operand. Otherwise, it returns False.
# Syntax: a > b
# Example:
a = 10
b = 5
c = a > b
# Output: Greater Than of a and b is True
print(f"Greater Than of {a = } and {b = } is", c)

# iv. Less Than Operator (<)
# The less than operator returns True if the left operand is less than the right operand. Otherwise, it returns False.
# Syntax: a < b
# Example:
a = 10
b = 5
c = a < b
# Output: Less Than of a and b is False
print(f"Less Than of {a = } and {b = } is", c)

# v. Greater Than or Equal To Operator (>=)
# The greater than or equal to operator returns True if the left operand is greater than or equal to the right operand. Otherwise, it returns False.
# Syntax: a >= b
# Example:
a = 10
b = 10
c = a >= b
# Output: Greater Than or Equal To of a and b is True
print(f"Greater Than or Equal To of {a = } and {b = } is", c)

# vi. Less Than or Equal To Operator (<=)
# The less than or equal to operator returns True if the left operand is less than or equal to the right operand. Otherwise, it returns False.
# Syntax: a <= b
# Example:
a = 10
b = 10
c = a <= b
# Output: Less Than or Equal To of a and b is True
print(f"Less Than or Equal To of {a = } and {b = } is", c)

# NOTE: The relational operators are also known as comparison operators. They are used to compare two operands. The result of a relational operator is always a boolean value (True or False).
# NOTE: In previous versions of Python there was another way to represent not equals operator. It was <> operator. But, it is no longer supported in Python 3.0 and above versions. So, we should use != operator instead of <> operator.
# Example:
# a = 20
# b = 30
# print(a <> b) # Output: SyntaxError: invalid syntax
```

```python
# Q 6. Illustrate Bitwise Operators
# The bitwise operators are used to perform bitwise operations on integer values.
# The following lists all the bitwise operators supported by Python.

# i. Bitwise AND Operator (&)
# The bitwise AND operator returns the result of bitwise AND operation on the two operands.
# Syntax: a & b
# Example:
a = 10     # In binary form: 0000 1010
b = 20     # In binary form: 0001 0100
c = a & b  # In binary form: 0000 0000
print(f"Bitwise AND of {a = } , {b = } is", c)
# Output: Bitwise AND of a , b is 0

# NOTE: With multiple operands, the bitwise AND operator returns the result of bitwise AND operation on the first two operands.
# Then, it performs bitwise AND operation on the result and the third operand.
# It continues this process until all the operands are processed.
# Example:
a = 10 & 20 & 30 & 40
print(f"Bitwise AND of 10, 20, 30 , 40 is", a)
# Output: Bitwise AND of 10, 20, 30 , 40 is 0
# Trick: If there is atleast one 0 in the binary form of the operands, the result of bitwise AND operation is 0. Otherwise, the result is 1.

# ii. Bitwise OR Operator (|)
# The bitwise OR operator returns the result of bitwise OR operation on the two operands.
# Syntax: a | b
# Example:
a = 10    # In binary form: 0000 1010
b = 20    # In binary form: 0001 0100
c = a | b  # In binary form: 0001 1110
# Output: Bitwise OR of a , b is 30
print(f"Bitwise OR of {a = } , {b = } is", c)

# NOTE: With multiple operands, the bitwise OR operator returns the result of bitwise OR operation on the first two operands.
# Then, it performs bitwise OR operation on the result and the third operand.
# It continues this process until all the operands are processed.
# Example:
a = 10 | 20 | 30 | 40
# Output: Bitwise OR of 10, 20, 30 , 40 is 62
print(f"Bitwise OR of 10, 20, 30 , 40 is", a)
# Trick: If there is atleast one true value, then the result is 1. If there is no true value, then the result is 0.

# iii. Bitwise XOR Operator (^)
# The bitwise XOR operator returns the result of bitwise XOR operation on the two operands.
# Syntax: a ^ b
# Example:
a = 10    # In binary form: 0000 1010
b = 20    # In binary form: 0001 0100
c = a ^ b  # In binary form: 0001 1110
# Output: Bitwise XOR of a , b is 30
print(f"Bitwise XOR of {a = } , {b = } is", c)

# NOTE: With multiple operands, the bitwise XOR operator returns the result of bitwise XOR operation on the first two operands.
# Then, it performs bitwise XOR operation on the result and the third operand.
# It continues this process until all the operands are processed.
# Example:
a = 10 ^ 20 ^ 30 ^ 40
# Output: Bitwise XOR of 10, 20, 30 , 40 is 62
print(f"Bitwise XOR of 10, 20, 30 , 40 is", a)
# Trick: If number of true values is odd, then the result is 1. If number of true values is even, then the result is 0.
```

```python

# Q 8, 9 and 60. Illustrate Precedence Rules and Associativity

# i. PEMDAS - Arithmetic operations
print("Eg. 1: PEMDAS")
print(2 + 3 * 4 - 6 / 2 + 2 * 3 ** (2 + 1), "\n")  # Output: 29.0
# Explanation: 0. (2 + 1) = 3      # Grouped operations are evaluated first
#              1. 3 ** 3 = 27      # Next highest precedence is of exponentiation
#              2. 3 * 4 = 12, 6 / 2 = 3.0, 2 * 27 = 54  # Mul., Div. are of same precedence so evaluated from left to right (associativity)
#              3. 2 + 12 - 3.0 + 54 = 65.0  # Finally Add. and Sub.
# Result is a float because one of the operands is a float

# ii. Binding or Parenthesised Expression, list, set and dict Display, Subscription, slicing, call, attribute reference
print("Eg. 2: Grouping and subsription")
print([1, 2, 3][0] + {"key": 3}["key"] + sum((1, 2, 3, 4,
      5, 6)[1:4]) - min({3, 4, 7, -1}), "\n")  # Output: 14
# Explanation: 0. [1, 2, 3], {"key": 3}, (1, 2, 3, 4, 5, 6) and {3, 4, 7, -1} are e+valuated # Parenthesiszed expressions
#              1. [1, 2, 3][0] = 1 ; {"key": 3}["key"] = 3 ;
#                 (1, ..., 6)[1:4] = (2, 3, 4) => sum((1,...,6)[1:4]) = 9 ;
#                 min({3,..,-1}) = -1 are evaluated # Subscription, slicing and call
#              2. 1 + 3 + 9 - (-1) = 14 # Addition and Subtraction

# iii. Unary +, - and Bitwise Operators
print("Eg. 3: Unary +, - and Bitwise Operators")
print(32 >> 2 << +5 + -2 + ~1 | 3 ^ 8 & 31, "\n")  # Output: 27
# Explanation: 0. +5 = 5 ; -2 = -2 ; ~1 = -2 are evaluated.
#              1. 5 - 2 - 2 = 1 is evaluated
#              2. 32 >> 2 = 8 => 8 << 1 = 16 is evaluated from left to right as >> and << have equal precedence
#              3. 8 & 31 = 8
#              4. 3 ^ 8 = 11
#              3. 16 | 11 = 27

# iv. Comparisons and Membership and Identity testing, Boolean Operators
# Test values
a = {1, 2, 3}
b = {True, False}
c = 256
d = 256
e = 12
n = -1
print("Eg. 4: Comparisons and Membership and Identity testing, Boolean Operators")
print(False or n > c and e and not False or c >=
      d in b or n and True in b or c != n, "\n")  # Otuput: True
# Explanation: 0. n > c = False ; c >= d in b is evaluated as c >= d and d in b = True and False = False ; True in b = True ; c != n = True
#              1. not False = True
#              2. False and 12 and True = False ; n and True is evaluated as -1 and True = True
#              3. False or False or False or True or True = True

# v. Assignment operations
print("Eg. 5: Assignment Operations")
x = 2
y = 4

x += y - 2
print(f"{x=}")  # Output: 4
# Explanation: 0. x += y - 2 is evaluated from right to left. So y - 2 is evaluated as 2 and then x += 2 updates x to be 4
# The Assignment Operations (Also called Walrus assignment Operator) - New in python 3.8
# More about it - https://peps.python.org/pep-0572/#examples
(x := 10 + 20)  # Note: The paranthesis around it is necessary as mentioned - https://peps.python.org/pep-0572/#exceptional-cases
print(f"{x=}\n")  # Output: 30
# Explanation: 0. 10 + 20 is evaluated as 30 and then x := 30 assigns 30 to x

# vi. Associativity
print("Eg. 6: Associativity")
# Output: 0.25 # Explanation: As //, /, % and * have equal precedence, associativity comes into play.
print(11 // 2 / 2 * 2.5 % 2)
#                              11 // 2 / 2 * 2.5 % 2 is evaluated from left to right. 11 // 2 = 5 => 5 / 2 = 2.5 => 2.5 * 2.5 = 6.25 => 6.25 % 2 = 0.25

print(2 + 3 - 7)  # Output: -2 # Explanation: Associativity from left to right
# Output: 256 # Explanation: Associativity from right to left -> 2 ** 2 ** 3 is evaluated as 2 ** 3 = 8 => 2 ** 8 => 256
print(2 ** 2 ** 3)
```

```python

# Q 31. Write a Python program to find Strong Number
def fact(n):
    p = 1
    for i in range(2, n + 1):
        p *= i
    return p


n = int(input("Enter a positive integer: "))
original = n
s = 0
while n != 0:
    d = n % 10
    s += fact(d)
    n //= 10

if s == original:
    print(original, "is a Strong Number")
else:
    print(original, "is not a Strong Number")
```

```python
# Q 71. Implement recursive Linear Search
def linear_search_r(arr, size, val):
    if size == 0:
        return -1

    if arr[size - 1] == val:
        return size - 1
    else:
        return linear_search_r(arr, size - 1, val)


a = [1, 2, 0, 3, 9, 4, 7]
size = len(a)
val = 3
print(linear_search_r(a, size, val))     # Output: 3
```

```python

# Q 73. Implement Recursive Binary Search


def binary_search_r(arr, val, low, high):
    if low > high:
        return -1
    else:
        mid = (low + high) // 2
        if arr[mid] == val:
            return mid
        elif arr[mid] > val:
            return binary_search_r(arr, val, low, mid - 1)
        else:
            return binary_search_r(arr, val, mid + 1, high)


a = [2, 5, 7, 18, 20, 30]
low = 0
high = len(a) - 1
val = 20
print(binary_search_r(a, val, low, high))
```

```python

# Q 74. Implement Iterative Binary Search


def binary_search_i(arr, val, low, high):
    while low != high:
        mid = (low + high) // 2
        if arr[mid] == val:
            return mid
        elif arr[mid] > val:
            high = mid - 1
        else:
            low = mid + 1

    return -1


a = [2, 5, 7, 18, 20, 30]
low = 0
high = len(a) - 1
val = 20
print(binary_search_i(a, val, low, high))
```

```python

# Q 75. Implement Bubble Sort


def bubble_sort(arr, size):
    for i in range(size):
        for j in range(size - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


a = [3, 5, 2, 10, 4, 1, 7]
size = len(a)
print(bubble_sort(a, size))
```

```python

# Q 76. Implement Selection Sort


def selection_sort(arr, size):
    for i in range(size):
        min_ = i
        for j in range(i, size):
            if arr[j] < arr[min_]:
                min_ = j

        if min_ != i:
            arr[i], arr[min_] = arr[min_], arr[i]

    return arr


a = [3, 5, 2, 10, 4, 1, 7]
size = len(a)
print(bubble_sort(a, size))
```

```python

# Q 77. Implement Matrix multiplication


def matrix_mul(A, B):
    if len(A[0]) != len(B):
        raise Exception(
            f"Invalid Matrix sizes for Multiplication : {len(A)} x {len(A[0])} and {len(B)} x {len(B[0])} -> {len(A[0])} != {len(B)}")
    else:
        m1, n1 = len(A), len(A[0])     # Matrix 'A' is of order m1 x n1
        m2, n2 = len(B), len(B[0])     # Matrix 'B' is of order m2 x n2
        # The reslut of A x B is the matrix of order m1 x n2. List comprehensions used instead of [...] * n to prevent aliasing
        res = [[0 for _ in range(n2)] for _ in range(m1)]

        for i in range(m1):
            for j in range(n2):
                s = 0
                for k in range(n1):
                    x = A[i][k]
                    y = B[k][j]
                    s += x * y
                res[i][j] = s

        return res


a = [[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6], [5, 6, 7]]  # Order 5 x 3
b = [[3, 2, 1, 4], [0, 3, 2, 5], [1, 5, 5, 7]]              # Order 3 x 4
# b = [[3, 2, 1], [0, 3, 2]]                                # Order 2 x 3 : Invalid Matrix sizes for Multiplication
print(matrix_mul(a, b))
```

```python

# Q 78. Implement Matrix Addition


def matrix_add(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        raise Exception(
            f"Invalid Matrix sizes for Addition : {len(A)} x {len(A[0])} and {len(B)} x {len(B[0])}")
    else:
        res = [[0 for _ in range(len(A[0]))] for _ in range(len(A))]
        for i in range(len(A)):
            for j in range(len(A[0])):
                res[i][j] = A[i][j] + B[i][j]

        return res


a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]   # Order 3 x 3
b = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]   # Order 3 x 3
# b = [[1, 2, 3], [2, 3, 5]]            # Order 2 x 3 : Invalid Matrix sizes for Addition
print(matrix_add(a, b))
```

```python
# Q 91. Implement linked list and basic operations (Using Classes)
# A linked list contains elements which are not stored in contiguous memory locations.
# They contain a reference to the next element in the list

# Implementing a Singly Linked List
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

  def __str__(self):
    return f"Node<data={self.data}, next={'Node<...>' if self.next is not None else None}>"


class LinkedList:
  def __init__(self):
    self.head = None

  # Displaying the LinkedList
  def __repr__(self):
    temp = self.head
    text = "LinkedList["
    while temp is not None:
      text += str(temp.data)
      if temp.next is not None:
        text += " -> "
      temp = temp.next
    text += "]"
    return text

  def __str__(self):
    temp = self.head
    text = "["
    while temp is not None:
      text += str(temp.data)
      if temp.next is not None:
        text += " -> "
      temp = temp.next
    text += "]"
    return text

  # Add a new item at front
  def insert_at_head(self, val):
    new_head = Node(val)
    new_head.next = self.head
    self.head = new_head
    return new_head

  # Add a new item after specified node
  def insert_after(self, prev_node, val):
    new_item = Node(val)
    new_item.next = prev_node.next
    prev_node.next = new_item
    return new_item

  # Add a new item at end
  def append(self, val):
    tail = Node(val)
    last = self.head
    if last is None:
      self.head = tail
    else:
      while last.next is not None:
        last = last.next
      last.next = tail
    return tail

  # Searching through the list
  def search(self, val):
    pos = 0
    temp = self.head
    while temp is not None:
      if temp.data == val:
        return temp, pos
      temp = temp.next
      pos += 1

  # Removing head
  def remove_head(self):
    if not self.head:
      return
    new_head = self.head.next
    self.head = new_head

  # Removing at given position (similar to index)
  def remove_at(self, pos):
    if pos == 0:
      return self.remove_head()
    prev = self.head
    for i in range(pos - 1):
      if prev == None:
        return
      prev = prev.next
    next = prev.next.next if prev.next is not None else None
    prev.next = next

  def remove_node(self, val):
    temp = self.head
    if temp.data == val:
      return self.remove_head()
    while temp.next is not None:
      if temp.data == val:
        prev.next = temp.next
        break
      prev = temp
      temp = temp.next

  def remove_after(self, prev_node):
    to_remove = prev_node.next
    if to_remove is None:
      return
    after = to_remove.next
    prev_node.next = after

  def remove_tail(self):
    if self.head is None:
      return
    if self.head.next is None:
      self.head = None
      return
    prev = self.head
    while prev.next.next is not None:
      prev = prev.next
    prev.next = None

  def reverse(self):
    prev = None
    curr = self.head
    while curr is not None:
      next = curr.next
      curr.next = prev
      prev = curr
      curr = next
    self.head = prev

  def sort(self):
    if self.head is None:
      return
    curr = self.head
    while curr.next is not None:
      next = curr.next
      while next is not None:
        if curr.data > next.data:
          curr.data, next.data = next.data, curr.data
        next = next.next
      curr = curr.next

  def __len__(self):
    temp = self.head
    count = 0
    while temp is not None:
      count += 1
      temp = temp.next
    return count


# Creating a new LinkedList
l_list = LinkedList()
print("Instantiate an empty linked list")
print(l_list)

# Adding elements
# i. To end
l_list.append(2)
print("\nAppending a node with value 2")
print(l_list)

# ii. At Head
print("\nInserting at head a node with value 7")
prev_node = l_list.insert_at_head(7)
print(l_list)

# iii. After specified node
print("\nInserting after a given node")
l_list.insert_after(prev_node, 3)
l_list.insert_after(prev_node, 11)
l_list.insert_after(prev_node, 4)
print(l_list)

# Removing elements
# i. Removing head
print("\nRemoving Head")
l_list.remove_head()
print(l_list)

# ii. Removing a node with specified value
print("\nRemoving node with specified value")
l_list.remove_node(3)
print(l_list)

# iii. Removing a node at specified position
print("\nRemoving node at specified position")
l_list.remove_at(1)
print(l_list)

head = l_list.head

# iv. Removing a node after specified node
print("\nRemoving node after a specific node")
l_list.remove_after(head)
print(l_list)

# Repopulating the linked list
print("\nRepopulating list to experiment more commands")
l_list.append(1)
l_list.append(4)
l_list.append(9)
l_list.append(16)
print(l_list)

# Searching for an item by value
print("\nSearching for a node by value")
node, pos = l_list.search(4)
print(node, "at pos.", pos)

# Reversing the linked list
print("\nReversing the linked list")
l_list.reverse()
print(l_list)

# Sorting the linked list
print("\nSorting the linked list")
l_list.sort()
print(l_list)

# Length of the linked list
print("\nLength of the linked list")
print(len(l_list))
```

```python
# Q 93. Write a program to demonstrate + and * operators on sequences.
# On sequences, + and * operators are overloaded to concatenate and repeat the sequences respectively.
# The + operator concatenates two sequences to produce a new sequence. The * operator repeats a sequence to produce a new sequence.
# Example:
l1 = [1, 2, 3]
l2 = [4, 5, 6]
print(f"{l1 + l2 = }")  # Output: l1 + l2 = [1, 2, 3, 4, 5, 6]
print(f"{l1 * 3 = }")  # Output: l1 * 3 = [1, 2, 3, 1, 2, 3, 1, 2, 3]

s1 = "Hello "
s2 = "World"
print(f"{s1 + s2 = }")  # Output: s1 + s2 = 'Hello World'
print(f"{s1 * 4 = }")  # Output: s1 * 4 = 'Hello Hello Hello Hello '

t1 = (1, 2, 3)
t2 = 4, 5, 6
print(f"{t1 + t2 = }")  # Output: t1 + t2 = (1, 2, 3, 4, 5, 6)
print(f"{t1 * 3 = }")  # Output: t1 * 3 = (1, 2, 3, 1, 2, 3, 1, 2, 3)

# NOTE: Sets and Dictionaries do not support + and * operators.

# In Technical Terms: The + operator is overloaded to concatenate two sequences. The * operator is overloaded to repeat a sequence.
```

```python
# Q 94. Write a program in python to determine the unique words and their frequency in a given string
# NOTE: To split the string at all delimiters after spaces, we use nested list comprehension
string = input("Enter the sentence / string:\n")
words = [z for x in string.lower().split(" ") for y in x.split("-")
         for z in y.split(".") if z.strip() != ""]
visit_track = [0] * len(words)
frequency = {}
for i in range(len(words)):
    if visit_track[i] == -1:
        continue
    else:
        word_count = 0
        cur_word = words[i]
        # Logic to find no. of instances of the same word in a later part. You can even use words.count(...)
        for j in range(i, len(words)):
            if words[j] == cur_word:
                word_count += 1
                visit_track[j] = -1
        frequency[words[i]] = frequency.setdefault(words[i], 0) + word_count
        visit_track[i] = -1

print("----------------------------")
print("|     Word     | Frequency |")
print("----------------------------")
for word, count in frequency.items():
    print(f"| {word:<13}|{count:^11}|")
print("----------------------------")
```

```python

# Q 96. Illustrate the usage of tell() and seek() functions
from io import UnsupportedOperation
#
# fp.seek() and fp.tell() are helpful to read specific parts of the open file
# fp.seek() helps us set the current stream position i.e., the number of characters moved till now
# fp.tell() gives the current stream position
#
# Syntax:
# fp.seek(target, whence)
# target is an integer representing number of characters to move through
# whence can have 3 values and represents the relative location to use:
#    0 : From beginning
#    1 : From current position  -> NOTE: For using values 1, 2 the file must be opened in 'b' mode. Exception is fp.seek(0, 2)
#    2 : From end of file
#
# fp.tell()
# returns the current stream position i.e, the number of characters read till now.
try:
    with open("text.txt", "br") as f:
        print("Stream position initially ->", f.tell())
        f.read(5)
        print("Stream position after f.read(5) ->", f.tell())
        f.seek(20)
        print("Stream position after f.seek(20) ->", f.tell())
        # Reading 5 characters from 20th position in file which was set as stream position using seek
        print("\tText from position 20", f.read(5), sep="\n")
        f.seek(20, 1)
        # Reading 5 characters from 25th pos.
        print("\tText from position 25", f.read(5), sep="\n")
        f.seek(-10, 2)  # to set stream pos to 10th position from end of file
except FileNotFoundError:
    print("Run the First Cell to create the dummy text file named 'text.txt' and then try running this cell again")
except UnsupportedOperation:
    print("Open the file in 'b' -> binary mode to use non-zero end-relative and cur-relative seeks")
```
